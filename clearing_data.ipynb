## All of function

```python
#增加標題
def get_title(soup):
    return soup.select('span')[0].text

#新增地址
def get_address(soup):
    lst = []
    lst.append(soup.select('span.addr')[0].text)
    return lst

#新增價錢
def get_price(soup):
    return soup.select('i')[0].text

#房東提供的家具
def hasFurn(soup):
    lst = []
    for each in soup.find_all('span',class_=re.compile('hometools')):
        lst.append(each.find_parent().text)
    return lst

#房東提供的設備
def hasEqui(soup):
    lst = []
    for each in soup.find_all('span',class_=re.compile('fixture')):
        lst.append(each.find_parent().text)
    return lst

#新增格局
def get_pattern(soup):
    pattern = {}
    for ii in soup.select('li'):
        # bookmark
        tmp = str(ii)
        try:
            if 'class' not in tmp:
                kk, vv = re.split(':|：',ii.text.replace(' ',''))
                pattern[kk.replace('\xa0','')] = vv.replace('\xa0','')  
        except:
            print(url)
    return pattern
#內文
def get_context(soup):
        context = {}
        i= 0
        for ii in soup.select('li.clearfix > div'):

            if i ==0 :
                kk = ii.text.strip()

            else:
                context[kk] = ii.text.replace('：', '')
            i = 1 if i == 0 else 0
        return context
#取得時間
def get_time(getTime):
        return getTime
    
        
```
